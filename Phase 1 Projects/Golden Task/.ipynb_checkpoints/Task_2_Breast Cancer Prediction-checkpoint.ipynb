## 1. Import required libraries

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

## 2. Load the dataset

```python
data = pd.read_csv('breast_cancer.csv')
data
```

### Dataset Description

The Data description is as follows:

1. diagnosis: The diagnosis of breast tissues (1 = malignant, 0 = benign) where malignant denotes that the disease is harmful
2. mean_radius: mean of distances from center to points on the perimeter
3. mean_texture: standard deviation of gray-scale values
4. mean_perimeter: mean size of the core tumor
5. mean_area: mean area of the core tumor
6. mean_smoothness: mean of local variation in radius lengths

## 4. Data Visualization

```python
sns.heatmap(data.corr(), annot=True, cmap = 'BrBG')
plt.show()
```

After visualizing this,
we can clearly say that 
* mean_perimeter and mean_radius is complete correlated(100%)
* mean_area and mean_radius(99%)
* mean_area and mean_perimeter(99%)

So, if we drop mean_perimeter column because it is highly correlated with mean_radius and we will also drop mean_area.

```python
data1 = data.drop(columns = ['mean_perimeter', 'mean_area'], axis = 1)
data1
```

```python
%%time
data.hist(bins=30, figsize=(18,12))
plt.show()
```

## 5. Spliting the dataset into training and testing

```python
X = data1.iloc[:,:-1].values
y = data1.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

## 6. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

## 7. Applying Machine Learning Models

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

## 2. K Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

## 3. Support Vector Machines (SVM)

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'linear')
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print(cm)
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.svm import SVC
kernel_svm = SVC(kernel = 'rbf')
kernel_svm.fit(X_train, y_train)
```

```python
y_pred = kernel_svm.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print(cm)
print(accuracy_score(y_test, y_pred))
```

##  4. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print(cm)
print(accuracy_score(y_test, y_pred))
```

## 5. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

## 6. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy')
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

## 7. Artificial Neural Network (ANN)

```python
import tensorflow as tf
import keras
from keras.layers import Dense, Dropout
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 3, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(units = 3, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
# compile the model
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', patience=5, verbose=1)
deep_history = model.fit(X_train, y_train, epochs=100, batch_size = 32, 
                         validation_data = (X_test, y_test), callbacks=[early_stop])
```

```python
def plot_loss(history):
    historydf = pd.DataFrame(history.history, index=history.epoch)
    plt.figure(figsize=(8, 6))
    historydf.plot(ylim=(0, historydf.values.max()))
    plt.title('Loss: %.3f' % history.history['loss'][-1])
plot_loss(deep_history)
plt.show()
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = (y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```

```python
df = pd.DataFrame({
    'Model Name': ['Logistic Regression', 'K Neighbors Classifier', 'Linear SVM', 'Kernal SVM', 'Naive Bayes', 'Decision Tree Classifier', 'Random Forest Classifier', 'ANN'],
    'Accuracy Score':  [91.23, 88.59, 88.59, 89.47, 91.23, 90.35, 89.47, 91.23]
})
df
```

```python
df = df.sort_values(by = ['Accuracy Score'], ascending = False)
df
```
